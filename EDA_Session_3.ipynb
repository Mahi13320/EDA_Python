## Map

- map is one method to convert categorical values to numeical values
- Take one Categorical column 
- make a dictionary with assign a number to each label
- for example case_status has two labels
  - Certified
  - Denied
- Assign 0 to certified , 1 to denied
- Create a dictionary labels as keys,number as values only
- {'Certified':0 , 'Denied':1}

```python
# step 1: read thw column
# step 2 : get the unique values
# step 3: make a dictionary
# step 4: apply the map
```

```python
vs_df['case_status'].unique()
```

```python
# I want apply the map method for all categorical columns
# we need to unique labels
# we need to get a numbers = len unique labels
# we need to create a dictionary
```

```python
unique_labels=sorted(vs_df['case_status'].unique())
list1=[i for i in range(len(unique_labels))]
dict1=dict(zip(unique_labels,list1))
dict1
vs_df['case_status']=vs_df['case_status'].map(dict1)
vs_df
```

```python
{key:value for key,value in zip(unique_labels,list1)}
```

```python
#cat_cols=vs_df.select_dtypes(include='object').columns
for i in cat_columns:
    unique_labels=sorted(vs_df[i].unique())
    list1=[i for i in range(len(unique_labels))]
    dict1=dict(zip(unique_labels,list1))
    dict1
    vs_df[i]=vs_df[i].map(dict1)
vs_df
```

# Label Encoder

- Labelencoder is a method to convert categorical data to numeical data
- labelencoder is a same way , how we did in the above
- The for loop map method we explored 
- But the labelEncoder also do the same 
- It is under **Sicket-learn** we usually called as **sklearn**
- under sklearn we have class called as preprocessing 
- Because currrently we are doing data preprocedings
- Under preprocessing class we have a method called as **LabelEncoder**
- Sklearn
    - preprocessing
    - LabelEncoder
- Any sklearn packkage has 3 steps
   - step 1: Read the package
   - step 2: Save the package
   - step 3 : Apply fit transform on data

```python
#step 1: Read the package
from sklearn.preprocessing import LabelEncoder
#step 2: Save the package
le=LabelEncoder()
#step 3 : Apply fit transform on data
vs_df['case_status']=le.fit_transform(vs_df['case_status'])
vs_df
```

```python
for i in cat_columns:
    vs_df[i]=le.fit_transform(vs_df[i])
vs_df
```

**np.where**

- using np.where also we can change the categorical data to numrical
- But np.where is only for binary condition
- np.where is same as if-else
- if condition is True: will apply True value
- if condition in False : will apply False value
- so np.where use for only binary labels
- A column should have only two unique labels

```python
con=vs_df['case_status']=='Certified'
vs_df['case_status']=np.where(con,0,1)
vs_df
```

**One hot encoder**

- One hot means one will be on ON , another will be OFF
- ON represents with 1
- OFF represents with 0

- for example case status has two unique labels
  - Certified
  - Denied
- One hot encoder cretes new column which is equal to number of unique labels
- for example case status will create two extra columns
  - case_status_Certified
  - case_status_Denied
